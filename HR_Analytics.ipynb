```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
df['Attrition']= label_encoder.fit_transform(df['Attrition'])
df['OverTime']= label_encoder.fit_transform(df['OverTime'])
df['Gender']= label_encoder.fit_transform(df['Gender'])
```

```python
df['Attrition']= df['Attrition'].replace({'N0':0, 'yes':1})
```

```python
df['Gender'].value_counts()
```

```python
df.head(2)
```

```python
cols= ['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus']
label_encoders ={}
for column in cols:
    label_encoders[column]= LabelEncoder()
    df[column]= label_encoders[column].fit_transform(df[column])
```

```python
df.head()
```

```python
df= df.drop(['Over18'], axis=1)
```

```python
# Model training

X= df.drop (['Attrition'], axis=1)
y= df['Attrition'].values
```

```python
pip install -U imbalanced-learn
```

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
r= RandomOverSampler()
X_t, y_t= r.fit_resample(X,y)
print (Counter (y_t))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split (X_t,y_t, test_size= 0.3, random_state= 2)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,roc_curve
DC= DecisionTreeClassifier(max_depth= 5)
DC.fit (X_train, y_train)
y_pred = DC.predict (X_test)
print (accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred, y_test))
print(roc_curve(y_pred, y_test))
```

```python
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict (X_test)
print (accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred, y_test))
print(roc_curve(y_pred, y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb= GradientBoostingClassifier()
gb.fit(X_train, y_train)
y_pred = gb.predict (X_test)
print (accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred, y_test))
print(roc_curve(y_pred, y_test))
```

```python
from sklearn.svm import SVC
svc= SVC()
svc.fit(X_train, y_train)
y_pred = svc.predict (X_test)
print (accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred, y_test))
print(roc_curve(y_pred, y_test))
```
