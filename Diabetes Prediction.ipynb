```python
# Importing Libraries for Prediction
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


```

```python
x= df.drop ('Outcome', axis=1)
y= df['Outcome']
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size= 0.2)
```

```python
model= LogisticRegression()   #Training the model
model.fit(x_train,y_train)
```

```python
prediction= model.predict(x_test)  #Fitting x train and y train into variable called  model
print(prediction)
```

```python
accuracy= accuracy_score(prediction, y_test)
print(accuracy)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn= KNeighborsClassifier()
knn.fit(x_train, y_train)
y_k= knn.predict(x_test)
print (accuracy_score(y_test,y_k))
print (accuracy)
```

```python
from sklearn.tree import DecisionTreeClassifier

ds= DecisionTreeClassifier()
ds.fit(x_train, y_train)
y_ds= ds.predict(x_test)
print (accuracy_score(y_test,y_ds))
print (accuracy)
```
